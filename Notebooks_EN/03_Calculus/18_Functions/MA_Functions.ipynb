##1. Draw in a single Geogebra notebook the following functions:

   - $g(x) = \sqrt{x}$
   - $h(x) = \frac{1}{x}$
   - $j(x) = \sin(x)$

Find value of all the above functions at $x = 2$.

1. For $f(x) = x^2$

$f(2) = 2^2 = 4$

2. For $g(x) = \sqrt{x}$

$g(x) = \sqrt{2} = 1.414$

3. For $h(x) = \frac{1}{x}$

$h(2) = \frac{1}{2} = 0.5$

4. For $j(x) = sin(x)$

$j(2) = sin(2) = 0.909$

2. Let $f(x) = 3x - 1$ and $g(x) = \sqrt{x}$. Find:
   - $f(g(x))$
   - $g(f(x))$
   - $f(f(x))$
   - $g(g(x))$

and visualize functions in a single Geogebra notebook.

1. $f(g(x))$

$f(g(x)) = f(\sqrt x) = 3(\sqrt x)-1 = 3\sqrt x-1$

2. $g(f(x))$

$g(f(x)) = g(3x-1) = \sqrt {3x-1}$

3. $f(f(x))$

$f(f(x)) = f(3x-1) = 3(3x-1) - 1 = 9x-3-1 = 9x-4$

4. $g(g(x))$

$g(g(x)) = g(\sqrt x) = \sqrt {\sqrt x} = x^\frac{1}{4}$

##3. Let $f(x) = e^x$ and $g(x) = \ln(x)$. Check: $f(g(x))$ and $g(f(x))$. What do you notice?

$f(g(x)) = f(\ln x) = e^{lnx}=x$

$g(f(x)) = g(e^x) = ln(e^x) = x$

Both compositions return x, meaning they are inverse functions.

##5. We have function $f=\{(1,7), (2,7), (3,11)\}$. Give inverse function $f^{-1}$.

However, note that for $f^-1$ to be a proper function, each input in $f^-1$ must have exactly one output. Here, 7 maps to both 1 and 2 which violates the definition of a function.
So, f^-1 is not a valid function.
